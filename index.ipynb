## Connecting to the Database

```python
# Your code here
import sqlite3
import pandas as pd
conn = sqlite3.connect('data.sqlite')
cur = conn.cursor()
```

## Display the names of all the employees in Boston 

Hint: join the employees and offices tables.

```python
# Your code here
cur.execute("""SELECT lastName, firstName, city
                FROM employees e
                JOIN offices o
                ON e.officeCode = o.officeCode
                WHERE city = 'Boston';""")
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
cur.execute("""SELECT o.officeCode, o.city, COUNT(e.employeeNumber) AS n_employees
               FROM offices AS o 
               LEFT JOIN employees AS e
               USING(officeCode)
               GROUP BY officeCode
               HAVING n_employees = 0;""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
df
```

```python
cur.execute("""SELECT COUNT(e.employeeNumber) AS Total_Employee,o.city,o.officeCode
                FROM offices o
                LEFT JOIN employees e
                USING(officeCode)
                GROUP BY officeCode
                HAVING Total_Employee = 0 """)
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here
cur.execute("""SELECT e.lastName,
                        e.firstName,
                        p.productCode,
                        p.productName,
                        p.productLine
                FROM employees e
                JOIN customers c
                ON e.employeeNumber = c.salesRepEmployeeNumber
                JOIN orders o
                USING(customerNumber)
                JOIN orderdetails od
                USING(orderNumber)
                JOIN products p
                USING(productCode);
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

```python
cur.execute("""SELECT   e.employeeNumber,
                        e.lastName,
                        e.firstName,
                        e.officeCode,
                        COUNT(p.productCode) as Total_Products
                FROM employees e
                JOIN customers c
                ON e.employeeNumber = c.salesRepEmployeeNumber
                JOIN orders o
                USING(customerNumber)
                JOIN orderdetails od
                USING(orderNumber)
                JOIN products p
                USING(productCode)
                GROUP BY e.employeeNumber
                ORDER BY Total_Products DESC;
                """)

df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```
